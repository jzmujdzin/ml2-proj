```python
# check for null values in df
df.isnull().sum()
```

```python
# train test split. 80% train, 20% test, y is class column. remap class to 0 and 1. use only two datasets train and test
from sklearn.model_selection import train_test_split
df['class'] = df['class'].map({'good': 0, 'bad': 1})
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# save train and test to csv, in folder /data
train.to_csv('data/classification_train.csv')
test.to_csv('data/classification_test.csv')
```
